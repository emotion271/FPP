```python
import numpy as np
import pandas as pd
import pickle as pkl
```

```python
path = "E:/FHN/data/polyvore/tuples_630"
```

```python
u_pos = pd.read_csv(path+'/tuples_train_posi')
```

```python
u_pos[:10]
```

```python
file = open(path+'/visual_feat.pkl', 'rb')
visual_vec = pkl.load(file)
```

```python
def read_image_list(image_list_fn):
    """Return image list for each fashion category.

    the image name of n-th item in c-th category in image_list[c][n]
    """
    files = image_list_fn
    return [open(fn).read().splitlines() for fn in files]
```

```python
image_list_fn = [path+'/image_list_top', path+'/image_list_bottom', path+'/image_list_shoe']
```

```python
image_list = read_image_list(image_list_fn)
```

```python
for i in range(630):
    uid = i
    df = u_pos[u_pos['user']==uid]
    #print(df)
    l = len(df)
    user_pos = [visual_vec[image_list[0][df.iloc[k,1]]] + visual_vec[image_list[1][df.iloc[k,2]]] + visual_vec[image_list[2][df.iloc[k,3]]] for k in range(len(df))]
    user_pos_np = np.array(user_pos)
    #print(user_pos_np)
    #print(user_pos_np.shape)
    f_mean = np.mean(user_pos_np,axis=0)
    #print(f_mean)
    #print(f_mean.shape)
    matrix_u = user_pos_np-f_mean
    #print(matrix_u.shape)
    arr = np.sum(matrix_u**2, axis=1)
    #print(arr)
    '''
    arr_cov = np.cov(user_pos_np)
    #print(arr_cov.shape)
    arr = np.diagonal(arr_cov)
    #print(arr_diag)
    '''
    idx = np.argsort(-arr)[:10]
    #print(idx)
    #break
    if i==0:
        user_his = df.iloc[idx]
    else:
        new = df.iloc[idx]
        user_his = pd.concat([user_his,new])
```
